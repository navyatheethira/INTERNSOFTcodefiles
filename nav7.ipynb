**Universal Functions:Index Alignment**

```python
area=pd.Series({'New York':1234567,'Tesla':7865948,'California':5602355}, name='area')
population=pd.Series({'California':7654766,'Tesla':7895436,'Swiz':1875439}, name='population')
print(area)
population
```

```python
population/area
```

```python
area.index.union(population.index)
```

**Combining Data Overlap**

```python
siri_a=pd.Series([np.nan,2.5,np.nan,3.5,4.5,np.nan],
                 index=['f','i','f','a','n','a'])
siri_b=pd.Series(np.arange(len(siri_a),dtype=np.float64),
                 index=['f','i','f','a','n','a'])
```

```python
siri_a
```

```python
siri_b
```

```python
pd.Series(np.where(pd.isnull(siri_a),siri_b,siri_a),index=siri_a.index)
```

```python
siri_a.combine_first(siri_b)
```
