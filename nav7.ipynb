**Combining Data Overlap**

```python
siri_a=pd.Series([np.nan,2.5,np.nan,3.5,4.5,np.nan],
                 index=['f','i','f','a','n','a'])
siri_b=pd.Series(np.arange(len(siri_a),dtype=np.float64),
                 index=['f','i','f','a','n','a'])
```

```python
siri_a
```

```python
siri_b
```

```python
pd.Series(np.where(pd.isnull(siri_a),siri_b,siri_a),index=siri_a.index)
```

```python
siri_a.combine_first(siri_b)
```
